# Generate Groups

```python
from datetime import datetime, timedelta
import random

# List of sample usernames
usernames = [
    "danielanderson", "conniejohnson", "cathy65", "davidjones", 
    "wendylane", "taylormiddleton", "cpaul", "huffmancarmen", "stephen25"
]

# Coupon-related message templates
coupon_messages = [
    "Just used my {company} coupon for {discount}. It worked!",
    "Got a {discount} off at {company}, anyone else tried this?",
    "Does anyone have a valid coupon code for {company}?",
    "Planning to use this {company} coupon soon, it's {discount}.",
    "I found a {company} coupon for {discount}, sharing it here.",
    "Redeemed my {discount} off at {company}, such a great deal!",
    "Is this {company} coupon still valid for {discount}?",
    "I need a good {company} coupon for {discount}.",
    "Shared a {company} coupon for {discount} earlier, did anyone try?",
    "Best coupon I've found so far is {discount} at {company}.",
]

# Function to generate a random date string in the format yyyy-MM-dd
def generate_random_date():
    today = datetime.today()
    random_days = random.randint(-100, 100)  # Random date within +/- 100 days
    random_date = today + timedelta(days=random_days)
    return random_date.strftime("%Y-%m-%d")

# Function to generate a coupon-related message
def generate_coupon_message(sender):
    companies = ["Amazon", "eBay", "Target", "Walmart", "Groupon", "Best Buy"]
    discounts = ["10% off", "15% off", "Buy one, get one free", "$20 off", "50% off", "$10 cashback"]
    
    message_template = random.choice(coupon_messages)
    message_text = message_template.format(
        company=random.choice(companies),
        discount=random.choice(discounts)
    )
    
    message = {
        "sender": sender,
        "data": message_text,
        "timestamp_string": generate_random_date()
    }
    return message

# Function to generate a group with name, admins, users, and coupon messages, sorted by date
def generate_group(group_name):
    admin_count = random.randint(1, 2)  # 1-2 admins per group
    user_count = random.randint(1, 3)   # 3-5 users per group
    
    admins = random.sample(usernames, admin_count)
    users = random.sample([u for u in usernames if u not in admins], user_count) + admins
    
    # Generating random coupon-related messages
    message_count = random.randint(5, 10)  # 5-10 messages per group
    messages = [generate_coupon_message(random.choice(admins + users)) for _ in range(message_count)]
    
    # Sort the messages by the timestamp_string (old to new)
    messages = sorted(messages, key=lambda x: datetime.strptime(x['timestamp_string'], "%Y-%m-%d"))
    
    group = {
        "name": group_name,
        "admins": admins,
        "users": users,
        "messages": messages,
        "picture": "\"\""
    }
    return group

# Example: Generating 3 groups
groups = [generate_group(f"Secret Coupons_{i}") for i in range(1, 4)]

# Print the generated groups as JSON
import json
print(json.dumps(groups, indent=4))
```
